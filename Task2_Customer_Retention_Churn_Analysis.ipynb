```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt

print("Libraries loaded successfully!")
```

```python
df = pd.read_csv("Telco-Customer-Churn.csv")

print(df.shape)
df.head()
```

```python
df.info()
```

```python
df.isnull().sum()
```

```python
df['Churn'].value_counts()
```

```python
import matplotlib.pyplot as plt

df['Churn'].value_counts().plot(kind='bar')

plt.title('Customer Churn Distribution')
plt.xlabel('Churn')
plt.ylabel('Number of Customers')
plt.show()
```

```python
import pandas as pd
import matplotlib.pyplot as plt

pd.crosstab(df['gender'], df['Churn']).plot(kind='bar')

plt.title('Churn by Gender')
plt.xlabel('Gender')
plt.ylabel('Count')
plt.show()
```

```python
pd.crosstab(df['Contract'], df['Churn']).plot(kind='bar')

plt.title('Churn by Contract Type')
plt.xlabel('Contract')
plt.ylabel('Count')
plt.show()
```

```python
pd.crosstab(df['InternetService'], df['Churn']).plot(kind='bar')

plt.title('Churn by Internet Service')
plt.xlabel('Internet Service')
plt.ylabel('Count')
plt.show()
```

```python
pd.crosstab(df['SeniorCitizen'], df['Churn']).plot(kind='bar')

plt.title('Churn by Senior Citizen')
plt.xlabel('Senior Citizen (0=No, 1=Yes)')
plt.ylabel('Count')
plt.show()
```

```python
plt.figure(figsize=(8,5))

df['tenure'].hist(bins=20)

plt.title('Customer Tenure Distribution')
plt.xlabel('Tenure (Months)')
plt.ylabel('Number of Customers')

plt.show()
```

```python
plt.figure(figsize=(8,5))

df['MonthlyCharges'].hist(bins=20)

plt.title('Monthly Charges Distribution')
plt.xlabel('Monthly Charges')
plt.ylabel('Number of Customers')

plt.show()
```

### Objective

### Key Insights

- Month-to-month customers show highest churn.
- Fiber optic customers have higher churn rates.
- Senior citizens are more likely to churn.
- New customers churn more frequently.
- Long-term contracts improve retention.

### Recommendations

- Promote annual subscriptions.
- Improve fiber service experience.
- Strengthen onboarding programs.
- Reward loyal customers.
- Offer personalized retention plans.
